**CODE TO DISPLAY 100 OR 100 EMPLOYEES DATA**

```python
import random #import random because it allow to have your desire uncountable number of data
import json #import json brcause to display data in dictionary or json format

names = ["maliha", "meheen", "hafsa", "manahil", "zara", "mehr", "aqsa", "kinza"]
departments = ["HR", "Software Engineering", "Marketing", "Sales", "Accounts"]
skills = ["html", "css", "JavaScript", "python", "ai engineer", "data science", "next js"]

employees = []

for i in range(1, 101):
    experiences = []
    for _ in range(random.randint(1, 3)):  # Each employee has 1 to 3 experiences
        experiences.append({
            "Company_Id": random.choice(["C001", "C002", "C003", "C004","C005","C006","C007","C008","C009","C010"]),
            "Age": random.randint(20, 60),
            "Department": random.choice(departments),
            "Salary": f"Rs.{random.randint(40000, 100000)}/-"
        })

    employee = {
        "ID": f"Emp{i}",
        "Name": random.choice(names),
        "Department": random.choice(departments),
        "Salary": f"Rs.{random.randint(40000, 100000)}/-",
        "Skills": f"{skills}",  # Multiple skills
        "Experience": experiences  # Multiple experiences
    }

    employees.append(employee)

# display details of employees
print(json.dumps(employees, indent=2))
```

**SAME CODE BY USING FUNCTIONS**

```python
import random
import json

# Sample data
names = ["maliha", "meheen", "hafsa", "manahil", "zara", "mehr", "aqsa", "kinza"]
departments = ["HR", "Software Engineering", "Marketing", "Sales", "Accounts"]
skills = ["html", "css", "JavaScript", "python", "ai engineer", "data science", "next js"]
company_ids = ["C001", "C002", "C003", "C004", "C005", "C006", "C007", "C008", "C009", "C010"]

# Function to generate random experiences
def generate_experiences():
    experiences = []
    for _ in range(random.randint(1, 3)):  # Each employee has 1 to 3 experiences
        experience = {
            "Company_Id": random.choice(company_ids),
            "Age": random.randint(20, 60),
            "Department": random.choice(departments),
            "Salary": f"Rs.{random.randint(40000, 100000)}/-"
        }
        experiences.append(experience)
    return experiences

# Function to generate a single employee
def generate_employee(emp_id):
    employee = {
        "ID": f"Emp{emp_id}",
        "Name": random.choice(names),
        "Department": random.choice(departments),
        "Salary": f"Rs.{random.randint(40000, 100000)}/-",
        "Skills":f"{skills}",  # Select 24 random skills
        "Experience": generate_experiences()
    }
    return employee

# Function to generate all employees
def generate_employees(count):
    employees = []
    for i in range(1, count + 1):
        employee = generate_employee(i)
        employees.append(employee)
    return employees

# Function to display employees in JSON format
def display_employees(employees):
    print(json.dumps(employees, indent=2))

# Call for function
employees = generate_employees(100)
display_employees(employees)

```
